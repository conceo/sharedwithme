```python
def multimed(a,n):
    
    global multi
    x = numpy.median([multi[i,a].value for i in range(1,n+1)])
    return(x)

def multimean(a,n):
    
    global multi
    x = numpy.mean([multi[i,a].value for i in range(1,n+1)])
    return(x)
```
