```python
import pandas
import numpy 
import ipywidgets as widgets
from ipywidgets import Layout, GridspecLayout, AppLayout
```

```python
blue_bg = "<style>.blue input { background-color:#66B3E9 !important; }</style>"
light_blue_bg = "<style>.light_blue input { background-color:#E1F3FF !important; }</style>"
light_grey_bg = "<style>.light_grey input { background-color:#BBCBD7 !important; }</style>"
```

```python
# Anzahl Vergleichsunternehmen
n = 4

# Festlegung Bewertungsmaßstab Multiplikatorverfahren
Bew = 0.15

inputvec = [1,2,3,4,5,6,7,8,9]
```

```python
# Eingabefeld
multi = GridspecLayout(n+4,7, column_size ='50px')

head = ['Vergleichsunternehmen','EV in MEUR','EV/Umsatz','EV/EBIT','EV/EBITDA','KGV']
row = ['Median', 'Mittelwert', 'Sektor Median']
for j in range(6):
    multi[0,j] = widgets.Text(head[j], disabled = True)
    multi[0,j].add_class('blue')
    
    for i in range(1,n+4):
        multi[i,j] = widgets.FloatText()
       
        
for i in range(1,n+1):
    multi[i,0] = widgets.Text('')
    multi[i,0].add_class('light_blue') 

for i in range(n+1,n+4):
    multi[i,0] = widgets.Text(row[i-(n+1)], disabled = True)
    multi[i,0].add_class('light_grey') 


def update1(change):
    multi[5,1] = widgets.FloatText(numpy.median([multi[i,1].value for i in range(1,n+1)]))
    multi[6,1] = widgets.FloatText(numpy.mean([multi[i,1].value for i in range(1,n+1)]))

def update2(change):
    multi[5,2] = widgets.FloatText(numpy.median([multi[i,2].value for i in range(1,n+1)]))
    multi[6,2] = widgets.FloatText(numpy.mean([multi[i,2].value for i in range(1,n+1)]))

def update3(change):
    multi[5,3] = widgets.FloatText(numpy.median([multi[i,3].value for i in range(1,n+1)]))
    multi[6,3] = widgets.FloatText(numpy.mean([multi[i,3].value for i in range(1,n+1)]))

def update4(change):
    multi[5,4] = widgets.FloatText(numpy.median([multi[i,4].value for i in range(1,n+1)]))
    multi[6,4] = widgets.FloatText(numpy.mean([multi[i,4].value for i in range(1,n+1)]))

def update5(change):
     multi[5,4] = widgets.FloatText(numpy.median([multi[i,5].value for i in range(1,n+1)]))
     multi[6,4] = widgets.FloatText(numpy.mean([multi[i,5].value for i in range(1,n+1)]))



# OBSERVE
multi[1,1].observe(update1, 'value')
multi[2,1].observe(update1, 'value')
multi[3,1].observe(update1, 'value')
multi[4,1].observe(update1, 'value')

multi[1,2].observe(update2, 'value')
multi[2,2].observe(update2, 'value')
multi[3,2].observe(update2, 'value')
multi[4,2].observe(update2, 'value')

multi[1,3].observe(update3, 'value')
multi[2,3].observe(update3, 'value')
multi[3,3].observe(update3, 'value')
multi[4,3].observe(update3, 'value')

multi[1,4].observe(update4, 'value')
multi[2,4].observe(update4, 'value')
multi[3,4].observe(update4, 'value')
multi[4,4].observe(update4, 'value')

multi[1,5].observe(update5, 'value')
multi[2,5].observe(update5, 'value')
multi[3,5].observe(update5, 'value')
multi[4,5].observe(update5, 'value')

# Bei Wahl von n - Wie geschickt erzeugen ohne for-Schleife??


# KALIBRIERUNGEN
outs = GridspecLayout(7,5)

outhead = ['','Umsatz','EBIT','EBITDA','KGV']
outrow = ['Zielunternehmen', 'Enterprise Value (Median)', 'Enterprise Value (Sektor)', 'Equity Bridge', 'Equity Value (Median)', 'Equity Value (Sektor)']

for j in range(5):
    outs[0,j] = widgets.Text(outhead[j], disabled = True)
    outs[0,j].add_class('blue')

    for i in range(1,7):
        outs[i,j] = widgets.FloatText()

for i in range(1,7):
    outs[i,0] = widgets.Text(outrow[i-1], disabled = True)
    outs[i,0].add_class('light_grey') 


# Zielunternehmen
outs[1,1] = widgets.FloatText(inputvec[1], disabled = True)
outs[1,2] = widgets.FloatText(inputvec[2], disabled = True)
outs[1,3] = widgets.FloatText(inputvec[3], disabled = True)
outs[1,4] = widgets.FloatText(inputvec[4], disabled = True)

# Equity Bridge
outs[4,1] = widgets.FloatText(inputvec[7], disabled = True)
outs[4,2] = widgets.FloatText(inputvec[7], disabled = True)
outs[4,3] = widgets.FloatText(inputvec[7], disabled = True)


def updatebig(change):
    # Enterprise Value (Median) 
    # - Umsatz 
    outs[2,1] = widgets.FloatText(inputvec[1]*multi[n+1,1].value)
    # - EBIT  
    outs[2,2] = widgets.FloatText(inputvec[2]*multi[n+1,2].value)
    # - EBITDA
    outs[2,3] = widgets.FloatText(inputvec[3]*multi[n+1,3].value)

    # Enterprise Value (Sektor)
    #  - Umsatz
    outs[3,1] = widgets.FloatText(inputvec[1]*multi[n+3,1].value)
    # - EBIT  
    outs[3,2] = widgets.FloatText(inputvec[2]*multi[n+3,2].value)
    # - EBITDA
    outs[3,3] = widgets.FloatText(inputvec[3]*multi[n+3,3].value)

    # Equity Value (Median)
    #  - Umsatz
    outs[5,1] = widgets.FloatText((outs[2,1].value + outs[4,1].value)*(1-Bew))
    # - EBIT  
    outs[5,2] = widgets.FloatText((outs[2,2].value + outs[4,2].value)*(1-Bew))
    # - EBITDA
    outs[5,3] = widgets.FloatText((outs[2,3].value + outs[4,3].value)*(1-Bew))
    # - Gewinn
    outs[5,4] = widgets.FloatText((outs[1,4].value)*(multi[n+1,5].value)*(1-Bew))

    # Equity Value (Sektor)
    #  - Umsatz
    outs[6,1] = widgets.FloatText((outs[3,1].value + outs[4,1].value)*(1-Bew))
    # - EBIT  
    outs[6,2] = widgets.FloatText((outs[3,2].value + outs[4,2].value)*(1-Bew))
    # - EBITDA
    outs[6,3] = widgets.FloatText(((outs[3,3].value + outs[4,3].value)*(1-Bew)))
    # - Gewinn
    outs[6,4] = widgets.FloatText((outs[1,4].value)*(multi[n+3,5].value)*(1-Bew))


# OBSERVE
multi[1,1].observe(updatebig, 'value')
multi[2,1].observe(updatebig, 'value')
multi[3,1].observe(updatebig, 'value')
multi[4,1].observe(updatebig, 'value')

multi[1,2].observe(updatebig, 'value')
multi[2,2].observe(updatebig, 'value')
multi[3,2].observe(updatebig, 'value')
multi[4,2].observe(updatebig, 'value')

multi[1,3].observe(updatebig, 'value')
multi[2,3].observe(updatebig, 'value')
multi[3,3].observe(updatebig, 'value')
multi[4,3].observe(updatebig, 'value')

multi[1,4].observe(updatebig, 'value')
multi[2,4].observe(updatebig, 'value')
multi[3,4].observe(updatebig, 'value')
multi[4,4].observe(updatebig, 'value')

multi[1,5].observe(updatebig, 'value')
multi[2,5].observe(updatebig, 'value')
multi[3,5].observe(updatebig, 'value')
multi[4,5].observe(updatebig, 'value')


grid1 = [widgets.HTML(blue_bg),widgets.HTML(light_blue_bg), widgets.HTML(light_grey_bg), multi]
grid2 = [widgets.HTML(blue_bg),widgets.HTML(light_grey_bg), outs]
```

```python
box1 = widgets.VBox(grid1)
box2 = widgets.VBox(grid2)
widgets.VBox([box1, box2])
```
